### Parcel Distance from Fresno City

This notebook adds the city_dist feature to the parcel dataset.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

import geof
```

```python
df = pd.read_csv("../joined_combined_filtered_mined_soil_water_df_td_13_4.csv")
```

```python
def city_dist(lat, long):
        
    dist = geof.geo_dist(36.747778, -119.7725, lat, long)

        
    return dist
```

```python
df["city_dist"] = df.apply(lambda x: city_dist(x["lat"], x["lng"]), axis=1)
```

```python
df.sort_values(by = "city_dist")
```
